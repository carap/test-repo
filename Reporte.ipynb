# Carga de archivos

```python
vmidssic = sqlContext.read.parquet(pathvmidssic)
```

```python
vmidshst = sqlContext.read.parquet(pathvmidshst)
```

```python
arbol = sqlContext.read.format('com.databricks.spark.csv')\
                  .option("delimiter", ",")\
                  .options(header='true',inferschema='true')\
                  .load(pathArbol_productos)
```

```python
sicoco = vmidssic.select('country_id','entity_id','interaction_sequence_id','process_time_date','start_date',
                         when(col('debit_date')=='0001-01-01','1900-01-01').otherwise(col('debit_date')).alias('debit_dateII'),
                         'service_type','service_name','subservice_type','subservice_name','transaction_status_type',
                         'commission_branch_id','commission_branch_name','commission_branch1_id',
                         'commission_banking_type','commission_currency_id','commission_account_id',
                         'commission_incumbent_name','branch_account_id','current_branch_account_name',
                         'branch_account1_id','service_banking_type','operation_currency_id','account_id',
                         'social_form_name','contract_id','total_operations_number',c('transaction_amount').alias("transaction_amount_sic"),
                         'fare_amount','exchange_currency_amount','commission_amount','vat_commission_amount',
                         'total_commission_amount','customer_type','rejected_operations_desc','customer_id',
                         'group_id','commerce_id','gl_account_id','charge_type','source_branch_id',
                         'upper_branch_id',c('banking_name').alias('banking_name_sic'),c('load_date').alias("load_date_sic")
                        )
```

# Joins

```python
sic_arb = sicoco.join(arbol,[sicoco.service_type == arbol.SERV_SICOCO,
                             sicoco.subservice_type == arbol.SUBSERV_SICOCO],
                        how = "inner"
                       )
```

```python
sic_arb.count()
```

```python
sic_arb_hst = sic_arb.join(vmidshst,vmidshst.function_channel_type == sic_arb.FUNCION,how = "inner")
```

```python
sic_arb_hst.count()
```

# Reporte

```python
reporte1 = sic_arb_hst.where((c('channel_id')==13) & (c('transaction_amount') >0) & (c('load_date') == '2018-01-11'))\
           .groupBy('SERVICIO','SUBSERVICIO','FUNCION',c('DESC_FUNCION').alias('DESCRIPCION_FUNCION'))\
           .agg(sum('transaction_id').alias("NUMERO_OPERACIONES"),sum('transaction_amount').alias("MONTO_OPERACION"))
```

```python
tabla = reporte1
```

```python
df = tabla.toPandas()
table = ff.create_table(df)
py.iplot(table, filename='jupyter/table1')
```

```python
trace = go.Bar(x=df.SUBSERVICIO,
               y=df.MONTO_OPERACION,
               name='Montos',
               marker=dict(color='#A2D5F2'))

data = [trace]

layout = go.Layout(title="Montos por subservicio",
               xaxis=dict(title='Subservicios'),
               yaxis=dict(title='Montos'))

fig = go.Figure(data=data, layout=layout)

py.iplot(fig, filename='jupyter/styled_bar')
```

```python
trace2 = go.Bar(x=df.SUBSERVICIO,
                y=df.NUMERO_OPERACIONES,
                name='Operaciones',
                marker=dict(color='#59606D'))

data2 = [trace2]

layout2 = go.Layout(title="Operaciones por subservicio",
                    xaxis=dict(title='Subservicios'),
                    yaxis=dict(title='Operaciones'))

fig2 = go.Figure(data=data2, layout=layout2)

py.iplot(fig2, filename='jupyter/styled_bar')
```
